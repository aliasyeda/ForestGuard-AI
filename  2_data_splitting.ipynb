```python
# 2_data_splitting.ipynb
"""
Customer Churn Analysis - Data Splitting
Task 2: Divide the data into training (80%) and testing (20%) sets
for model training and evaluation, ensuring a representative split.
"""

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import os

print("="*80)
print("TASK 2: DATA SPLITTING")
print("80% Training / 20% Testing Split with Stratification")
print("="*80)

# Step 1: Load preprocessed data
print("\n📂 STEP 1: LOADING PREPROCESSED DATA")
print("-"*50)

try:
    X = pd.read_csv('data/processed/X_processed.csv')
    y = pd.read_csv('data/processed/y_processed.csv')['Churn']
    print(f"✅ Data loaded successfully!")
    print(f"   Features shape: {X.shape}")
    print(f"   Target shape: {y.shape}")
except:
    print("❌ Error: Preprocessed data not found!")
    print("Please run 1_data_preprocessing.ipynb first")
    exit()

# Step 2: Check class distribution before split
print("\n📊 STEP 2: CLASS DISTRIBUTION BEFORE SPLIT")
print("-"*50)

churn_count = (y == 1).sum()
no_churn_count = (y == 0).sum()
total = len(y)

print(f"Total samples: {total}")
print(f"Churn (Yes): {churn_count} ({churn_count/total*100:.1f}%)")
print(f"No Churn (No): {no_churn_count} ({no_churn_count/total*100:.1f}%)")

# Step 3: Split data with stratification
print("\n✂️ STEP 3: SPLITTING DATA (80% Train, 20% Test)")
print("-"*50)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,           # 20% for testing
    random_state=42,         # For reproducibility
    stratify=y               # Maintain class distribution
)

print(f"✅ Data split completed!")
print(f"\n📊 Split Results:")
print(f"   Training set: {X_train.shape[0]} samples ({X_train.shape[0]/total*100:.0f}%)")
print(f"   Testing set: {X_test.shape[0]} samples ({X_test.shape[0]/total*100:.0f}%)")

# Step 4: Verify class distribution in splits
print("\n📊 STEP 4: VERIFYING CLASS DISTRIBUTION")
print("-"*50)

print(f"\nTraining set distribution:")
train_churn = (y_train == 1).sum()
train_no_churn = (y_train == 0).sum()
print(f"   Churn: {train_churn} ({train_churn/len(y_train)*100:.1f}%)")
print(f"   No Churn: {train_no_churn} ({train_no_churn/len(y_train)*100:.1f}%)")

print(f"\nTesting set distribution:")
test_churn = (y_test == 1).sum()
test_no_churn = (y_test == 0).sum()
print(f"   Churn: {test_churn} ({test_churn/len(y_test)*100:.1f}%)")
print(f"   No Churn: {test_no_churn} ({test_no_churn/len(y_test)*100:.1f}%)")

# Step 5: Save split data
print("\n💾 STEP 5: SAVING SPLIT DATA")
print("-"*50)

X_train.to_csv('data/processed/X_train.csv', index=False)
X_test.to_csv('data/processed/X_test.csv', index=False)
y_train.to_csv('data/processed/y_train.csv', index=False)
y_test.to_csv('data/processed/y_test.csv', index=False)

print(f"✅ Training features saved: data/processed/X_train.csv")
print(f"✅ Testing features saved: data/processed/X_test.csv")
print(f"✅ Training target saved: data/processed/y_train.csv")
print(f"✅ Testing target saved: data/processed/y_test.csv")

# Step 6: Data split summary
print("\n📊 STEP 6: SPLIT SUMMARY")
print("-"*50)

split_summary = pd.DataFrame({
    'Dataset': ['Training', 'Testing', 'Total'],
    'Samples': [len(X_train), len(X_test), total],
    'Churn %': [
        f"{(y_train==1).mean()*100:.1f}%",
        f"{(y_test==1).mean()*100:.1f}%",
        f"{churn_count/total*100:.1f}%"
    ],
    'No Churn %': [
        f"{(y_train==0).mean()*100:.1f}%",
        f"{(y_test==0).mean()*100:.1f}%",
        f"{no_churn_count/total*100:.1f}%"
    ]
})

print(split_summary.to_string(index=False))

print("\n" + "="*80)
print("✅ TASK 2 COMPLETED: DATA SPLITTING")
print("="*80)
print("\n📝 Summary of Actions:")
print("1. ✅ Loaded preprocessed features and target")
print("2. ✅ Verified class distribution before split")
print("3. ✅ Split data 80/20 with stratification")
print("4. ✅ Verified balanced distribution in splits")
print("5. ✅ Saved training and testing sets")
print("\n🎯 Key Insight: Stratification ensures the test set represents")
print("   the original class distribution, crucial for imbalanced datasets.")
print("\n🚀 Ready for next step: Feature Selection (Task 3)")
```
